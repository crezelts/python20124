```python
#  

import pandas as pd
import matplotlib.pyplot as plt

# CSV  
data = pd.read_csv('/content/drive/MyDrive/seoul_inoculation.csv', encoding='euc-kr')

#   datetime  
data[''] = pd.to_datetime(data[''])

# 2021 2022  
data_2021 = data[data[''].dt.year == 2021]
data_2022 = data[data[''].dt.year == 2022]

# 2021  
plt.figure(figsize=(10, 6))
plt.plot(data_2021[''], data_2021['1  '], label='1  ', marker='o')
plt.plot(data_2021[''], data_2021['2  '], label='2  ', marker='o')
plt.plot(data_2021[''], data_2021['3  '], label='3  ', marker='o')
plt.title('   (2021) - 1, 2, 3 ')
plt.xlabel('')
plt.ylabel(' ')
plt.legend()
plt.rc('font', family='NanumBarunGothic')
plt.tight_layout()
plt.show()
```

```python
#  

# 2022  
plt.figure(figsize=(10, 6))
plt.bar(data_2022[''], data_2022['1  '], label='1  ', width=3)
plt.bar(data_2022[''], data_2022['2  '], label='2  ', width=3)
plt.bar(data_2022[''], data_2022['3  '], label='3  ', width=3)
plt.title('   (2022) - 1, 2, 3 ')
plt.xlabel('')
plt.ylabel(' ')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#   plt.plot() plt.fill_between()  

import seaborn as sns

# CSV  
data = pd.read_csv('/content/drive/MyDrive/seoul_deth.csv', encoding='euc-kr')

#   datetime  
data[''] = pd.to_datetime(data[''])

#    
data['  '] = data[''].cumsum()

#  
plt.figure(figsize=(12, 6))
sns.lineplot(x='', y='  ', data=data, color='red', linewidth=2)
plt.fill_between(data[''], data['  '], color="salmon", alpha=0.4)

#     
plt.title('      (2020 ~ 2022)', fontsize=16)
plt.xlabel('', fontsize=14)
plt.ylabel('  ', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.rc('font', family='NanumBarunGothic')
plt.grid()
plt.show()
```

```python
#   Seaborn sns.lineplot()  

import numpy as np
import seaborn as sns

# CSV  
data = pd.read_csv('/content/drive/MyDrive/seoul_deth.csv', encoding='euc-kr')

#   datetime  
data[''] = pd.to_datetime(data[''])

#   datetime  
data[''] = pd.to_datetime(data[''])

#    
data['  '] = np.cumsum(data[''])  # numpy cumsum    

# Seaborn   
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='', y='  ', color='red', linewidth=2)
plt.fill_between(data[''], data['  '], color='salmon', alpha=0.4)

#     
plt.title('    (2020 ~ 2022)', fontsize=16)
plt.xlabel('', fontsize=14)
plt.ylabel('  ', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid()
plt.show()
```

```python
# 선 그래프

import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일을 읽어오기
data = pd.read_csv('/content/drive/MyDrive/seoul_inoculation.csv', encoding='euc-kr')

# 날짜 열을 datetime 형식으로 변환
data['날짜'] = pd.to_datetime(data['날짜'])

# 2021년과 2022년 데이터 필터링
data_2021 = data[data['날짜'].dt.year == 2021]
data_2022 = data[data['날짜'].dt.year == 2022]

# 2021년 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(data_2021['날짜'], data_2021['1차 누적 접종자수'], label='1차 누적 접종자', marker='o')
plt.plot(data_2021['날짜'], data_2021['2차 누적 접종자수'], label='2차 누적 접종자', marker='o')
plt.plot(data_2021['날짜'], data_2021['3차 누적 접종자수'], label='3차 누적 접종자', marker='o')
plt.title('서울특별시 중랑구 예방접종 진행상황(2021년) - 1차, 2차, 3차 접종')
plt.xlabel('날짜')
plt.ylabel('접종자 수')
plt.legend()
plt.rc('font', family='NanumBarunGothic')
plt.tight_layout()
plt.show()
```

```python
# 막대 그래프

# 2022년 그래프 그리기
plt.figure(figsize=(10, 6))
plt.bar(data_2022['날짜'], data_2022['1차 누적 접종자수'], label='1차 누적 접종자', width=3)
plt.bar(data_2022['날짜'], data_2022['2차 누적 접종자수'], label='2차 누적 접종자', width=3)
plt.bar(data_2022['날짜'], data_2022['3차 누적 접종자수'], label='3차 누적 접종자', width=3)
plt.title('서울특별시 중랑구 예방접종 진행상황(2022년) - 1차, 2차, 3차 접종')
plt.xlabel('날짜')
plt.ylabel('접종자 수')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 이 그래프는 plt.plot()과 plt.fill_between()을 사용해서 표현함

import seaborn as sns

# CSV 파일을 읽어오기
data = pd.read_csv('/content/drive/MyDrive/seoul_deth.csv', encoding='euc-kr')

# 날짜 열을 datetime 형식으로 변환
data['날짜'] = pd.to_datetime(data['날짜'])

# 누적 사망자 수 계산
data['누적 사망자 수'] = data['사망자수'].cumsum()

# 그래프 그리기
plt.figure(figsize=(12, 6))
sns.lineplot(x='날짜', y='누적 사망자 수', data=data, color='red', linewidth=2)
plt.fill_between(data['날짜'], data['누적 사망자 수'], color="salmon", alpha=0.4)

# 그래프 제목 및 레이블 설정
plt.title('서울특별시 중랑구 코로나 누적 사망자 수 (2020년 ~ 2022년)', fontsize=16)
plt.xlabel('날짜', fontsize=14)
plt.ylabel('누적 사망자 수', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.rc('font', family='NanumBarunGothic')
plt.grid()
plt.show()
```

```python
# 이 그래프는 Seaborn의 sns.lineplot()을 사용하여 시각화

import numpy as np
import seaborn as sns

# CSV 파일을 읽어오기
data = pd.read_csv('/content/drive/MyDrive/seoul_deth.csv', encoding='euc-kr')

# 날짜 열을 datetime 형식으로 변환
data['날짜'] = pd.to_datetime(data['날짜'])

# 날짜 열을 datetime 형식으로 변환
data['날짜'] = pd.to_datetime(data['날짜'])

# 누적 사망자 수 계산
data['누적 사망자 수'] = np.cumsum(data['사망자수'])  # numpy의 cumsum을 사용하여 누적 합계 계산

# Seaborn을 사용한 그래프 그리기
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='날짜', y='누적 사망자 수', color='red', linewidth=2)
plt.fill_between(data['날짜'], data['누적 사망자 수'], color='salmon', alpha=0.4)

# 그래프 제목 및 레이블 설정
plt.title('서울특별시 누적 사망자 수 (2020년 ~ 2022년)', fontsize=16)
plt.xlabel('날짜', fontsize=14)
plt.ylabel('누적 사망자 수', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid()
plt.show()
```
